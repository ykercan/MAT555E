For this question consider the [Car Evaluation Data Set](https://archive.ics.uci.edu/ml/datasets/Car+Evaluation) from UCI. Here is the [direct link](https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data) to the dataset.

Make [contingency tables](https://en.wikipedia.org/wiki/Contingency_table#:~:text=In%20statistics%2C%20a%20contingency%20table,%2C%20engineering%2C%20and%20scientific%20research.) of the columns (using [`crosstab`](https://pandas.pydata.org/docs/reference/api/pandas.crosstab.html) function from [pandas](https://pandas.pydata.org)) and figure out which pairs of columns are dependent and independent. Explain your result using statistical tests.

```python
import pandas as pd

dataset = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data",sep=',', header=None)
dataset
```

```python
a = pd.crosstab(dataset[0],dataset[1])
b = pd.crosstab(dataset[0],dataset[2])
c = pd.crosstab(dataset[0],dataset[3])
d = pd.crosstab(dataset[0],dataset[4])
e = pd.crosstab(dataset[0],dataset[5])
f = pd.crosstab(dataset[0],dataset[6])

g = pd.crosstab(dataset[1],dataset[2])
h = pd.crosstab(dataset[2],dataset[3])
i = pd.crosstab(dataset[3],dataset[4])
j = pd.crosstab(dataset[4],dataset[5])
k = pd.crosstab(dataset[5],dataset[6])

l = pd.crosstab(dataset[1],dataset[3])
m = pd.crosstab(dataset[2],dataset[4])
n = pd.crosstab(dataset[3],dataset[5])
o = pd.crosstab(dataset[4],dataset[6])

p = pd.crosstab(dataset[1],dataset[4])
r = pd.crosstab(dataset[2],dataset[5])
s = pd.crosstab(dataset[3],dataset[6])

x = pd.crosstab(dataset[1],dataset[5])
y = pd.crosstab(dataset[2],dataset[6])

z = pd.crosstab(dataset[1],dataset[6])
```

```python
from scipy.stats import chisquare

print(chisquare(a,axis=None))
print(chisquare(b,axis=None))
print(chisquare(c,axis=None))
print(chisquare(d,axis=None))
print(chisquare(e,axis=None))
print(chisquare(f,axis=None))
print(chisquare(g,axis=None))
print(chisquare(h,axis=None))
print(chisquare(i,axis=None))
print(chisquare(j,axis=None))
print(chisquare(k,axis=None))
print(chisquare(l,axis=None))
print(chisquare(m,axis=None))
print(chisquare(n,axis=None))
print(chisquare(o,axis=None))
print(chisquare(p,axis=None))
print(chisquare(r,axis=None))
print(chisquare(s,axis=None))
print(chisquare(x,axis=None))
print(chisquare(y,axis=None))
print(chisquare(z,axis=None))
```

The results show that the p values, that are equal to zero, containts columns that are dependent to each other.

```python
result = [a,b,c,d,e,f,g,h,i,j,k,l,m,n,o,p,r,s,x,y,z]

for i in result:
    if chisquare(i,axis=None).pvalue !=1 :
        print(chisquare(i,axis=None))
```
